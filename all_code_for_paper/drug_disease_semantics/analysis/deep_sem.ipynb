```python
def aggregate_deep_sem_votes(votes):
    temp = []
    for deep_sem, group in votes.groupby("deep_semantics"):
        conf_score = sum(group["_trust"])
        num_votes = len(group)
        
        temp.append([deep_sem, conf_score, num_votes])
        
    return pd.DataFrame(temp, columns = ["deep_semantic_choice", "conf_score", "num_votes"])
```

```python
def get_crowd_deep_sem(crowd_broad, raw_data):
    temp = []
    for unit_id, top_broad in crowd_broad.items():
        if top_broad in ["positive", "speculative"]:
            subset = raw_data.query("_unit_id == {0}".format(unit_id))
            subset = subset.query("broad_rel_type == '{0}'".format(top_broad))

            res = aggregate_deep_sem_votes(subset)
            
            res.insert(0, "unit_id", unit_id)
            
            res = res.sort(["conf_score"], ascending = False)
            
            temp.append(res)
            
    return pd.concat(temp)
```

```python
# get crowd ans:
crowd_broad = dict()
for unit_id, group in results.groupby("unit_id"):
    top_choice = group.iloc[0]["broad_rel_type"]
    
    crowd_broad[unit_id] = top_choice

crowd_deep = get_crowd_deep_sem(crowd_broad, raw_data)

top_crowd_deep = dict()
for unit_id, group in crowd_deep.groupby("unit_id"):
    top_crowd_deep[unit_id] = group.iloc[0]["deep_semantic_choice"]
```

```python
toby_broad = dict()
toby_deep = dict()
for unit_id, group in toby_resp.groupby("unit_id"):
    toby_broad[unit_id] = group.iloc[0]["toby_broad_reltype"]
    
    if toby_broad[unit_id] in ["positive", "speculative"]:
        toby_deep[unit_id] = group.iloc[0]["toby_deep_sem"]
```

```python
def compare_responses(group_a, group_b):
    """
    Broad semantics.
    Compares to find the unique work ids that groups a and b agreed upon or disagreed upon.
    """
    agree = set()
    disagree = set()
    
    # there is no way to evaluate work units which only one group worked on
    common_units = set(group_a.keys()) & set(group_b.keys())
    print "Common number of work units:", len(common_units)
    for unit_id in common_units:
        if group_a[unit_id] == group_b[unit_id]:
            agree.add(unit_id)
        else:
            disagree.add(unit_id)
            
    return (agree, disagree)
```

```python
agree, disagree = compare_responses(toby_deep, top_crowd_deep)
```

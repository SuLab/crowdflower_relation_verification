## Functions

```python
# get crowd ans:
crowd_broad = dict()
for unit_id, group in results.groupby("unit_id"):
    top_choice = group.iloc[0]["broad_rel_type"]
    
    crowd_broad[unit_id] = top_choice
```

Broad agreement between the three

```python
def compare_responses(group_a, group_b):
    """
    Broad semantics.
    Compares to find the unique work ids that groups a and b agreed upon or disagreed upon.
    """
    agree = set()
    disagree = set()
    
    # there is no way to evaluate work units which only one group worked on
    common_units = set(group_a.keys()) & set(group_b.keys())
    print "Common number of work units:", len(common_units)
    for unit_id in common_units:
        if group_a[unit_id] == group_b[unit_id]:
            agree.add(unit_id)
        else:
            disagree.add(unit_id)
            
    return (agree, disagree)
```

```python
toby_broad = dict()
for unit_id, group in toby_resp.groupby("unit_id"):
    toby_broad[unit_id] = group.iloc[0]["toby_broad_reltype"]
```

## Three way compariosn

```python
ct_agree, ct_disagree = compare_responses(toby_broad, crowd_broad)
```

```python
cg_agree, cg_disagree = compare_responses(crowd_broad, euadr_answer)
```

```python
tg_agree, tg_disagree = compare_responses(toby_broad, euadr_answer)
```

```python
all_agree = ct_agree & cg_agree & tg_agree
print len(all_agree)
```
